```python

import pickle
import re
import os
filepath = "./dravidian-codemix.txt"
```

```python
files = open(filepath, "r")
data = files.readlines()

def get_tokens(block_no):
  tokens = []
  subset = data[14000:] if block_no == 15 else data[(block_no -1)*1000: (block_no-1+3)*1000]
  for line in subset:
    temp = nltk.word_tokenize(re.sub(r'[^\w\s]', '', line))
    
    temp.append("\n")
    tokens.extend(temp)
  print("found", len(tokens), "tokens in block", block_no)
  return tokens
```

Here we are using words from nltk.corpus package. This is a good choice for small datasets. For larget datasets one can use the Enchant Words Corpus. 

The program tags textual datasets that contain atmost two languages only. Each end of a sentence can be identified with the <'\n'> token with tag NA. For tagging more than 2 languages, use DetectLanguage API.

```python
from nltk.corpus import words

def get_lang(res):
  
  try: 
     if res.lower() in words.words():
       ret = 'en' 
     elif res =='\n':  
       ret ='NA' 
     else:
       ret = 'ta'
     return ret
  except: return 'NA'

def get_annotations(tokens, start_index =0 , N=1000):
  end_index = len(tokens) 
  batch = tokens[start_index:end_index]
  print("detected language for tokens",start_index,"to",end_index)
  print("NEXT START INDEX : ",start_index+N) if end_index != len(tokens) else print("END OF BLOCK") 
  languages = [get_lang(res) for res in batch]
  
  return [(batch[i],languages[i]) for i in range(len(languages))]
```

```python
def driver(filename, block_no, start_index):
  
  tokens = get_tokens(block_no)
  annotations = get_annotations(tokens)
  print(annotations)
  pkl_filepath = os.path.join("./",filename)
  dataset = []
  if(os.path.isfile(pkl_filepath)):
    print("found existing pickle object")
    f = open(pkl_filepath,"rb")
    dataset = pickle.load(f)
    f.close()
  else: print("Creating new pickle object")
  dataset.extend(annotations)
  f = open(pkl_filepath,"wb")
  pickle.dump(dataset,f)
  f.close()
  print("Pickle object updated")
```

```python
driver("pickle.dat",4,1000)
```
